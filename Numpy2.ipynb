Сгенерируем выборку числе среднего объёма при помощи нормальной случайной величины numpy.random.normal (распределение Гаусса):

```python
data = np.random.normal(10, 5, 20000)
# data
```

Для первичного "взгляда"на эти числа можно отобразить их в виде гистограммы (как бы визуализация чатотного анализа с предварительнным округлением вниз):

```python
plt.hist(data, bins=20)
plt.show()
```

Нормальная случайная величина, распределённая по Гауссу, напоминает метание дротика в мишень. Мы целимся в 10-ку, но у нас также есть средняя некая ширина промаха.

Среднее арифметическое и среднеквадратическое отклонение идеально пожходят для описания подобного распределения:

```python
print(f"Среднее арифметическое: {data.mean()}")
print(f"Среднеквадратическое отклонение: {data.std()}")
```

Как видите, при начально заданных модельных характеристиках 10 и 5 соответственно, мы нашли их очень точно по ланной выборке.

медиана - самый важный квартиль. Это число, которое находится в середине этого набора, если его упорядочить по возростанию, то есть такое число, что половина из элементов набора не меньше него, а другая половина не больше.

Первый квартиль - это 0,25. Этоо число, меньше которого находится 1/4 чисел набора, а больше него 3/4 набора.

Третий квартиль - это 0,75 квартиль. Это число, меньше которого находится 3/4 чисел набора, а больше него 1/4 набора.

Второй квартиль не говорят, т.к. это получается то же самое, что медиана.

## #Создание многомерных массивов

```python
A = np.zeros(shape=(3, 5), dtype=np.int64)
print(A)
```

Также можно сделать reshape одномерного массива: 

```python
A = np.arange(0, 60).reshape(6, 10) # Первый индекс - высота, второй - ширина
print(A)
```

```python
np.zeros_like(A) # Создать массив по форме как данный, но состоящий из нулей
```
